```python
import pandas as pd
import seaborn as sns
```

```python
final_df = pd.read_csv("resources/Final_Merge.csv")
people_df = pd.read_csv("resources/pers_19.csv")
```

```python
#I was curious how often each region showed up in the dataset
region_count = final_df.groupby('REGIONNAME_x').count()
region_count
```

```python
makes = sns.histplot(
    data = final_df, y = "MAKENAME_x", hue = "MAKENAME_x", element = "bars", multiple = "layer",
    shrink = 1,  stat = 'count', thresh = "1", fill = True, legend = False)
makes.figure.set_figheight(11)
```

```python
weather = sns.histplot(data = final_df, x = "NUM_INJV", hue="VEH_ALCHNAME", binwidth = 1, discrete = True,
             binrange = (0,4), multiple="stack")
weather.figure.set_figheight(8)
```

```python
ages = sns.histplot(data = people_df, x = "AGE_IM", hue = "DRINKINGNAME", discrete = True, multiple = "stack",
                   binrange = (0, 100))
ages.figure.set_figwidth(12)
```
