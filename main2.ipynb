```python
# Which region of the US has the terrible drivers?

selected_data = new_vehicle_df[['CASENUM', 'MAKENAME', 'REGIONNAME']]

# Group data by region and count the number of accidents and vehicles involved in each region
region_stats = selected_data.groupby('REGIONNAME').agg({'CASENUM': 'count', 'MAKENAME': 'sum'})

# Sort regions by the number of accidents
sorted_regions = region_stats.sort_values(by='CASENUM', ascending=False)

# Display the region with the highest number of accidents
worst_driving_region = sorted_regions.index[0]
print("Region with awful drivers based on the number of accidents:")
print(worst_driving_region)
```

```python
# Does time influence the number of accidents
hourname_groupby = new_accident_df.groupby(["HOURNAME"])
max_hourname = hourname_groupby ["HOURNAME"].count()
max_hourname 
max_hourname_sorted = max_hourname .sort_values(ascending=False)
max_hourname_sorted.head()
```

```python
# Does weather influence the number of accidents
weather_groupby = new_accident_df.groupby(["WEATHERNAME"])
max_weather = weather_groupby["WEATHERNAME"].count()
max_weather
max_weather_sorted = max_weather.sort_values(ascending=False)
max_weather_sorted.head()
```

```python
# Does month influence the number of accidents
monthname_groupby = accident_df.groupby(["MONTHNAME"])
max_monthname = monthname_groupby ["MONTHNAME"].count()
max_monthname 
max_monthname_sorted = max_monthname.sort_values(ascending=False)
max_monthname_sorted.head()
```

```python
# Which Makes get into accidents more frequently? 

selected_data = new_vehicle_df['MAKENAME']

# Count the frequency of each vehicle make
make_freq = selected_data.value_counts()

# Sort makes by frequency in descending order
sorted_makes = make_freq.sort_values(ascending=False).head(10)

# Display the makes that get into accidents more frequently
print("Makes that get into accidents more frequently:")
print(sorted_makes)
```

```python
# Severity proportion and accident frequency for each region?
selected_data = new_vehicle_df[['REGIONNAME', 'DEFORMEDNAME']]

# Group data by region and severity, and count the number of accidents
region_severity_counts = selected_data.groupby(['REGIONNAME', 'DEFORMEDNAME']).size().unstack(fill_value=0)

# Calculate the total number of accidents for each region
region_total_accidents = region_severity_counts.sum(axis=1)

# Calculate the severity proportion for each region
region_severity_proportion = region_severity_counts.div(region_total_accidents, axis=0)

# Display the severity proportion and accident frequency for each region
print("Severity proportion and accident frequency for each region:")
print(region_severity_proportion)
```

```python
import csv
from pathlib import Path
# Additional new Data Validation library
from pydantic import BaseModel

# Define a Pydantic model for the data in the CSV file
class Vehicle(BaseModel):
    MAKENAME: str
    CASENUM: int
    

# Define the path to the CSV file
filePath = Path("NewVehicle.csv")

# Getting user input  CAR MAKE: Ford CASENUM- 201901182028; 201901182028; 201901174219
                    # CAR MAKE: Honda CASENUM - 201901176655; 201901180224
car_make = input('Please enter the car make: ')
case_number = input('Please enter the case number: ')

# Open the CSV file
with open(filePath, newline='') as csvfile:
    # Create a CSV reader
    reader = csv.DictReader(csvfile)
    
    # Iterate through each row in the CSV file
    for row in reader:
        # Create a Pydantic model instance for the current row
        vehicle = Vehicle(**row)
        
        # Check if the case number and car make match
        if car_make.lower() == vehicle.MAKENAME.lower() and case_number == str(vehicle.CASENUM):
            # Print the details of the vehicle
            print("Vehicle details:")
            for k, v in row.items():
                print(k, ':', v)
            break
    else:
        print('This number does not exist')
```
