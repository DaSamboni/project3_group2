```python
import pandas as pd
```

```python
vehicle_data = pd.read_csv("resources/veh_19.csv", low_memory = False)
people_data = pd.read_csv("resources/pers_19.csv", low_memory = False)
accident_data = pd.read_csv("resources/acc_19.csv", low_memory = False)
```

```python
#----------------------------------------------------------------#
```

```python
vehicle_df = pd.DataFrame(vehicle_data)
vehicle_df.head()
```

```python
people_df = pd.DataFrame(people_data)
people_df.head()
```

```python
accident_df = pd.DataFrame(accident_data)
accident_df.head()
```

```python
cols = [0, 1, 3, 5, 6, 7, 10, 11, 12, 15, 17, 21, 23, 25, 27, 29, 32, 34, 36, 38, 40,
        72, 74, 76, 78, 80, 82, 84, 86, 88, 89, 96, 98, 100, 102, 112, 114, 116, 118,
        120, 122, 124, 126, 138, 146]

new_vehicle_df = vehicle_df[vehicle_df.columns[cols]]
```

```python
new_vehicle_df.head()
```

```python

new_vehicle_df.to_csv("NewVehicle.csv", index=False, header=True)
```

```python
# Displays all the column
pd.set_option('display.max_columns', None)
accident_df.head(2)
```

```python
new_accident_df = accident_df[["CASENUM","STRATUMNAME","REGION","REGIONNAME","PSU","PJ","URBANICITYNAME","NUM_INJ","MONTH","YEAR","DAY_WEEKNAME",
                               "HOURNAME","HARM_EVNAME","ALCOHOLNAME","MAX_SEVNAME","MAN_COLLNAME","RELJCT2NAME",
                               "TYP_INTNAME","LGT_CONDNAME","WEATHERNAME"]]
new_accident_df.head()
```

```python
new_accident_df.to_csv("NewAccident.csv", index=False, header=True)
```

```python
# Displays all the column
pd.set_option('display.max_columns', None)
people_df.head(2)
```

```python
new_people_df = people_df[["CASENUM","VE_FORMS","VEH_NO","PER_NO","STRATUM","REGION","REGIONNAME","PSU","PJ","URBANICITYNAME","MONTHNAME","HOURNAME","MINUTE","HARM_EVNAME","MAN_COLLNAME",
                               "MAKENAME","MAK_MODNAME","MOD_YEARNAME","TOW_VEHNAME","ROLLOVERNAME","IMPACT1NAME","AGENAME","SEXNAME","PER_TYPNAME","INJ_SEVNAME",
                               "SEAT_POSNAME","AIR_BAGNAME","DRINKINGNAME","ALC_STATUSNAME","DRUGSNAME","HOSPITALNAME"]]
new_people_df.head()
```

```python
new_people_df.to_csv("NewPeople.csv", index=False, header=True)
```

```python
Merge_df = new_vehicle_df.merge(new_accident_df,how='inner',on='CASENUM').merge(new_people_df,how='inner',on='CASENUM')
```

```python
Merge_df.head()
```

```python
Merge2_df = Merge_df.drop_duplicates(subset="CASENUM")
```

```python
Merge2_df.head()
```

```python
Final_Merge_df = Merge2_df[["CASENUM", "VEH_NO_x", "REGIONNAME_x", "REGION_x", "HOURNAME_x", "MONTHNAME_x", "MAKENAME_x", "MAX_SEVNAME", "NUM_INJV", "VEH_ALCHNAME", "VSURCONDNAME"]]
```

```python
Final_Merge_df.head()
```

```python
Final_Merge_df.to_csv("resources/Final_Merge.csv")
```
